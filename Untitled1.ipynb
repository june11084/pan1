```python
import pandas as pd
```

```python
df_read = pd.read_csv('lrs_network_by_year.csv')
df_read[df_read['client_1'] == df_read['client_2']]
```

```python
df_loop = df_read[df_read['client_1'] == df_read['client_2']].groupby('client_2')['total_mv'].sum()/7
df_out = df_read[df_read['client_1'] != df_read['client_2']].groupby('client_1')['total_mv'].sum()/7
df_in = df_read[df_read['client_1'] != df_read['client_2']].groupby('client_2')['total_mv'].sum()/7
```

```python
df_out.loc[4]
```

```python
df_size = pd.concat([df_loop, df_in], axis=1).fillna(0).sum(axis=1)
df_size = df_size.fillna(0) - df_out.fillna(0)
df_size = df_size.fillna(0)
df_size.loc[df_size < 1] = 1
df_size = df_size.round(0).astype(int)
df_size.to_csv('sizes.csv')
```
